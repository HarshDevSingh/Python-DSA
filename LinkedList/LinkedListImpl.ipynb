```python
class Node:
    def __init__(self,data=None,next=None):
        self.data=data
        self.next=next
```

```python
class LinkedList:
    def __init__(self):
        self.head=None
        
    def get_length(self):
        length=0
        if self.head is None:
            return length
        itr=self.head
        while itr:
            length+=1
            itr=itr.next
        return length
        
    def print_ll(self):
        if self.get_length()<1:
            return "Linked list is empty"
        itr=self.head
        ll=""
        while itr:
            suffix="-->" if itr.next else ""
            ll+=str(itr.data)+suffix
            itr=itr.next
        return ll
            
    def insert_at_begining(self,data):
        node=Node(data,self.head)
        self.head=node
        return "success"
    
    def insert_at_end(self,data):
        itr=self.head
        while itr.next:
            itr=itr.next
        node=Node(data,itr.next)
        itr.next=node
        return "success"
    
    def insert_at_index(self,index,data):
        if index < 0 or index not in range(self.get_length()):
            return "out of range index was provided"
        if index ==0:
            self.insert_at_begining(data)
        if index == self.get_length()-1:
            self.insert_at_end(data)
        itr=self.head
        count=0
        while itr:
            if count==index-1:
                node=Node(data,itr.next)
                itr.next=node
                break
            count+=1
            itr=itr.next
        return "success"
    
    def remove_at_index(self,index):
        if index < 0 or index not in range(self.get_length()):
            return "out of range index was provided"
        
        if index==0:
            self.head=self.head.next
            return "Node deleted, successfully!"
        itr =self.head
        count=0
        while itr:
            if count == index-1:
                itr.next=itr.next.next
                break
            count +=1
            itr=itr.next
        return "Node deleted, successfully!"
    
    def bulk_insert(self, data_list=[], at_end=True):
        if len(data_list)<1:
            return "data list can not be empty, please provide valid list"
        for data in data_list:
            if at_end:
                self.insert_at_end(data)
            else:
                self.insert_at_begining(data)
        return "data inserted, successfully!"
```
