```python
def find_sum(lst):
    res=[]
    nums=lst
    ln= len(lst)
    
    for i in range(ln-1):
        if i>0 and nums[i]==nums[i-1]:
            continue
        l=i+1
        r=ln-1
        while l<r:
            total=nums[i]+nums[l]+nums[r]
            if total <0:
                l+=1
            elif total>0:
                r-=1
            else:
                res.append([nums[i],nums[l],nums[r]])
                while l<r and nums[l]==nums[l+1]:
                    l+=1
                while l<r and nums[r]==nums[r-1]:
                    r-=1
                l+=1
                r-=1
    return res
    
```
